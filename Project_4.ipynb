```python
#setup
import bs4 as bs
import urllib.request
import pandas as pd
SDG_Cat = []
SDG_Num = []
year = []
var_names = []
val_val = []
dic = {}
df = pd.DataFrame(columns = ['SDG_Category', 'SDG_Number', 'Year', 'Variable_Name', 'Variable_Value'])
address = 'https://sdg.data.gov/'
site = urllib.request.urlopen(address).read()
lxml = bs.BeautifulSoup(site, 'lxml')

#making a list of all the initial urls
links = []
links2 = []
for url in lxml.find_all('a'):
    links.append((url.get('href')))
del links[0:6]
links.remove("#")
links = [x[2:] for x in links]

for page in links:
    stor = []
    temp3 = ""
    lst = urllib.request.urlopen(address+page).read()
    lst = bs.BeautifulSoup(lst, 'lxml')
    for url in lst.find_all('a'):
        stor.append(url.get("href"))
    del stor[0:7]
    stor.remove("#")
    links2.extend(stor)
    
#making dict for number to category
    
    stor = [x[1:] for x in stor]
    temp3 = str(page[:-1])
    stor = stor[0]
    dic[stor[0:2]] = temp3
    
    
#take slash off links

links2 = [x[1:] for x in links2]

#Grab all cvs from cite

for item in links2:
    temp2 = []
    address2 = address+"data/indicator_"+item+'.csv'
    csvs = pd.read_csv(address2)
    year_lp = csvs.iloc[:,0]
    csvs = csvs.iloc[:,1:]
    var_names_lp = list(csvs)
    csvs = csvs.T
#add data to final chart

    val_val_lp = []
    #print(csvs)
# for each column in teh dataframe grabs data to list and then adds lists togther

    for column in csvs:
        lst2 = csvs.iloc[:,column]
        val_val_lp.extend(list(lst2))
    count = int(len(val_val_lp)/len(var_names_lp)) 
    var_names_lp = var_names_lp * (count)


#add var name data and val val data to chart
    
    var_names.extend(var_names_lp)
    val_val.extend(val_val_lp)
    year_lp = list(year_lp)
    count = int(len(val_val_lp)/len(year_lp))
    year_temp = []
    
#dulpicating years correctly
    
    for x in year_lp:
        temp = []
        temp.append(x)
        temp = temp * count
        year_temp.extend(temp)
        year_lp = year_temp
    year.extend(year_lp)
    
#adding the file number used    
    
    temp2.append(item)
    temp2 = temp2 * int(len(val_val_lp))
    SDG_Num.extend(temp2)
    
# final category setup
for key in SDG_Num:
    SDG_Cat.append(dic[key[0:2]])
    
# final chart construction

from IPython.display import display, HTML
df1 = pd.DataFrame({'SDG_Category': SDG_Cat})
df1 = df1.assign(SDG_Number=pd.Series(SDG_Num))
df1 = df1.assign(Year=pd.Series(year))
df1 = df1.assign(Variable=pd.Series(var_names))
df1 = df1.assign(Value=pd.Series(val_val))
pd.set_option('display.max_columns', None)  
display(df1)
```
